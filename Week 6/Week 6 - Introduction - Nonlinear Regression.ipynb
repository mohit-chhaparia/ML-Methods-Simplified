```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)
n = 200
x = np.linspace(-3 , 3 , n)
y_true = 0.5 * x ** 3 - x
y = y_true + np.random.normal(0 , 1.0 , n)
X = x.reshape(-1 , 1)
linear_model = LinearRegression()
linear_model.fit(X , y)
y_pred = linear_model.predict(X)
residuals = y - y_pred
window = 15
kernel = np.ones(window) / window
residual_smooth = np.convolve(residuals , kernel , mode = "same")
```

```python
plt.figure(figsize = (8.5 , 5.2))
plt.scatter(x , y , alpha = 0.7 , s = 35 , label = "Data")
plt.plot(x , y_true , linestyle = "--" , linewidth = 2.5 , label = "True nonlinear pattern")
plt.plot(x , y_pred , linewidth = 3 , label = "Linear baseline")
plt.title("Curved data can fool a straight-line model")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(alpha = 0.25)
plt.tight_layout()
plt.show()
```
