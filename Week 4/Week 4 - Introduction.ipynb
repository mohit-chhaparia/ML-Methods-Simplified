```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)

n = 200
X = np.random.uniform(-3 , 3 , n).reshape(-1 , 1)
true_w = 2.5
true_b = -1.0
noise = np.random.normal(0 , 2 , n)
y = true_w * X.flatten() + true_b + noise

model = LinearRegression()
model.fit(X , y)
y_pred = model.predict(X)
```

```python
plt.figure()
plt.scatter(X , y)
plt.plot(X , y_pred)
plt.title("Synthetic Dataset + Baseline Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
w_vals = np.linspace(-1 , 5 , 100)
loss_vals = []

for w in w_vals:
    y_hat = w * X.flatten()
    loss = np.mean((y - y_hat) ** 2)
    loss_vals.append(loss)

plt.figure()
plt.plot(w_vals , loss_vals)
plt.title("Loss vs Weight (1D Slice)")
plt.xlabel("Weight")
plt.ylabel("MSE Loss")
plt.show()
```
