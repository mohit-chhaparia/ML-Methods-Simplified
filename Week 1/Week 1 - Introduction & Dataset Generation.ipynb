```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
```

```python
rng = np.random.default_rng(42)
n = 200
```

```python
cats = rng.multivariate_normal(mean = [-1.0 , -1.0] , cov = [[1.0 , 0.3] , [0.3 , 1.0]] , size = n)
cats[:10]
len(cats)

dogs = rng.multivariate_normal(mean = [1.2 , 1.0] , cov = [[1.0 , -0.2] , [-0.2 , 1.0]] , size = n)
dogs[:10]
len(dogs)
```

```python
X = np.vstack([cats , dogs])
X[:10]
X.shape

y = np.hstack([np.zeros(n) , np.ones(n)])
y[:10]
y.shape
```

```python
print("X shape:" , X.shape , "| y mean (class balance):" , y.mean())
print("First 3 rows of X:\n" , X[:3])
```
